```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
import numpy as np
path="/home/PuMengYu/nnUNet_workspace/raw/Dataset004_LiverTumor/imagesTr/liver_0_0000.nii.gz"
img=sitk.ReadImage(path)#读取.nii/.nii.gz/.mha等格式
img_array=sitk.GetArrayFromImage(img)#将图像转换成numpy数组
#注意转numpy,轴顺序会变,sitk是(x,y,z),转成numpy是(z,y,x)
#获取元数据
print("获取像素间距:",img.GetSpacing())#获取像素间距
#这里的像素间距表示voxel在三个轴方向的物理尺寸,也就是每个像素在轴上的代表的实际距离
print("获取像素原点坐标:",img.GetOrigin())#获取像素原点坐标
print("获取像素方向:",img.GetDirection())#获取像素方向
#这里的X列(-1,0,0)意思是图像的x轴正轴是物理向左,
print("获取像素尺寸:",img.GetSize())#获取像素尺寸
# organ_mask = sitk.BinaryThreshold(
#         label, lowerThreshold=1, upperThreshold=2,
#         insideValue=1, outsideValue=0)
```
